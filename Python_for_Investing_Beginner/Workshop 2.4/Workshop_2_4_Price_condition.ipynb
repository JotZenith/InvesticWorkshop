```python
import pandas as pd
import yfinance as yf
```

```python
t0 = '2021-01-01'
t1 = '2022-11-30'
```

```python
data = yf.download('BTC-USD', start=t0, end=t1)
```

```python
data
```

```python
df = data.copy()
```

```python
df['Return'] = df['Close'].pct_change()
```

```python
df
```

```python
df[df.Close > 60000]
```

# ข้อที่ 2 Filter Return ที่มีค่ามากกว่า 0

```python
df[df['Return'] > 0]
```

# ข้อที่ 3 สร้าง Function แล้วให้ Print โดยมี Condition ที่ว่าหาก Return > 0 ให้ Print "Gain" หาก Return < 0 ให้ปริ้น "Loss"

```python
def ret_con(df):
  if (df > 0):
    print('Gain')

  elif (df < 0):
    print('Loss')
```

```python
for i in range(len(df.Close)):
  ret_con(df.iloc[i,-1])
```

# แถม

```python
sum1 = len(df.loc[(df['Return'] < 0)])
sum2 = len(df.loc[(df['Return'] > 0)])

sum = len(df.Return)

per1 = (sum1/sum *100)
per2 = (sum2/sum *100)

print("Number of expected loss are {0} times in {1} about {2:.2f}%".format(sum1,sum, per1))
print("Number of expected gain are {0} times in {1} about {2:.2f}%".format(sum2,sum, per2))
```
